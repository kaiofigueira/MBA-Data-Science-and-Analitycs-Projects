```python
import pandas as pd

# Exemplo de dataset
dados = {
    "cliente": ["Ana", "João", "Maria", "Pedro", "Lucas", "Carla", "Marcos", "Paula"],
    "idade": [25, 40, 35, 50, 22, 33, 41, 29],
    "cidade": ["SP", "RJ", "SP", "BH", "SP", "RJ", "BH", "SP"],
    "valor_compra": [200, 350, 400, 150, 120, 300, 500, 180]
}

df = pd.DataFrame(dados)

# --- Exemplo de análise descritiva ---
# Estatísticas básicas
print(df.describe())

# Média de valor de compra por cidade
print("\nMédia de compras por cidade:")
print(df.groupby("cidade")["valor_compra"].mean())

# Perfil dos clientes por cidade (idade média)
print("\nIdade média por cidade:")
print(df.groupby("cidade")["idade"].mean())

# Correlação entre idade e valor gasto
print("\nCorrelação idade x valor_compra:")
print(df[["idade", "valor_compra"]].corr())
```
